```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('twitchdata-update.csv')
```

```python
df.groupby('Channel')['Followers'].max() \
    .sort_values(ascending=False) \
    .head(20) \
    .sort_values() \
    .plot(kind='barh',
          figsize=(50, 20),
          title='Top followed streamers')
plt.show()
```
